```python
import os

import librosa
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

## Automatic Music Tune Features Extraction

```python
directories = ["blues", "classical", "country", "hiphop", "disco", "jazz", "pop", "reggae", "rock", "metal"]

df = pd.DataFrame()
pd_columns_list = ["genre", "zero_crossings", "cpectral_centroids_mean", "spectral_centroids_var", "spectral_rolloff_mean", "spectral_rolloff_var", "mfccs_mean", "mfccs_var", "chroma_frequencies_mean", "chroma_frequencies_var"]
df = pd.DataFrame(columns = pd_columns_list)

a = []

for i in directories:
    for filename in os.scandir(i):
        if not filename.name.startswith('.'):
            a.append(i + "\\" + filename.name)

for p in a:

    def get_info_from_a_tune(p):

        # 0/ Initializing variables
        row = []

        ## Loading audiofile
        x_music_array, sr_sample_rate = librosa.load(p)

        # 1/ getting genre
        genre = p.split('\\', 1)[0]
        row.append(genre)

        
        # 2/ zero_crossings
        zero_crossings = librosa.zero_crossings(x_music_array, pad = False)
        par_1 = sum(zero_crossings)
        row.append(par_1)

        
        # 3/ spectral_centroids
        spectral_centroids = np.mean(librosa.feature.spectral_centroid(x_music_array, sr = sr_sample_rate))
        spectral_centroids_var = np.var(librosa.feature.spectral_centroid(x_music_array, sr = sr_sample_rate))
        
        row.append(spectral_centroids)
        row.append(spectral_centroids_var)
    

        # 4/ spectral_rolloff
        spectral_rolloff = np.mean(librosa.feature.spectral_rolloff(x_music_array, sr = sr_sample_rate))
        spectral_rolloff_var = np.var(librosa.feature.spectral_rolloff(x_music_array, sr = sr_sample_rate))
        
        row.append(spectral_rolloff)
        row.append(spectral_rolloff_var)

        # 5/ mfccs
        mfccs = np.mean(librosa.feature.mfcc(x_music_array, sr = sr_sample_rate))
        mfccs_var = np.var(librosa.feature.mfcc(x_music_array, sr = sr_sample_rate))
        
        row.append(mfccs)
        row.append(mfccs_var)
    
        
        # 6/ chroma_frequencies
        chromogram = np.mean(librosa.feature.chroma_stft(x_music_array, sr = sr_sample_rate))
        chromogram_var = np.var(librosa.feature.chroma_stft(x_music_array, sr = sr_sample_rate))
        
        row.append(chromogram)
        row.append(chromogram_var)
        

        #7/ Fulfilling dataframe
        name = p.split('\\')
        df.loc[name[1]] = row #for each name of the audio - a row

        return df

    df = get_info_from_a_tune(p)
    
```

## Creating Dictionary for Encoder (to understand the answer given by a model)
        

```python
le = preprocessing.LabelEncoder()

le_fitted = le.fit_transform(directories)
inverted = le.inverse_transform(le_fitted)

list_1 = le_fitted.tolist()
list_2 = inverted.tolist()

di = {}

for i in list_1:
    for j in list_2:
        di[i] = j
        list_2.remove(j)
        break
```

```python
def getting_values_and_encoding1_column(dataset):
    le = preprocessing.LabelEncoder()
    le_fitted = le.fit_transform(dataset["genre"])
    
    dataset["genre"] = le.fit_transform(dataset["genre"])
    
    X = dataset.drop(columns = 'genre')
    y = dataset['genre']
    
    return X, y

X, y = getting_values_and_encoding1_column(df)
```

```python
def train_test(dataset):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 21) # 21, 42 & 0 are popular ones
    
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = train_test(df)
```

```python
def scalling(dataset):
    scaler = StandardScaler()

    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    return X_train_scaled, X_test_scaled, scaler

X_train_scaled, X_test_scaled, scaler = scalling(df)
```

# CNN algorithm

```python
def train_model(model, epochs, optimizer):
    batch_size = 128
    model.compile(optimizer=optimizer,
                 loss='sparse_categorical_crossentropy',
                 metrics='accuracy'
    )
    
    return model.fit(X_train_scaled, y_train, validation_data=(X_test_scaled, y_test), epochs=epochs,
                    batch_size=batch_size)
```

```python
model = Sequential([
    
    Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.2),
    
    Dense(64, activation='relu'),
    Dropout(0.2),
    
    Dense(10, activation='softmax'),
])
```

```python
model_history = train_model(model=model, epochs=600, optimizer='adam')
```

```python
test_loss, test_acc = model.evaluate(X_test_scaled, y_test, batch_size=128)
print()
print("Test loss is:", test_loss)
print("\nBest test accuracy is:", test_acc * 100)
```

## Genre Detecting Function - enter the name of a music tune in the end of the function

```python
def showing_genre_of_a_tune(p):
    
    #path = "" # with genres folder
    # 0/ Initializing variables
    row = []
    

    ## Loading audiofile
    x_music_array, sr_sample_rate = librosa.load(p) #path + name_or_just_name
    
    
    # 2/ zero_crossings (par_1)
    zero_crossings = librosa.zero_crossings(x_music_array, pad = False)
    par_1 = sum(zero_crossings)
    row.append(par_1)

    # 3/ spectral_centroids (par_2)
    spectral_centroids = np.mean(librosa.feature.spectral_centroid(x_music_array, sr = sr_sample_rate))
    spectral_centroids_var = np.var(librosa.feature.spectral_centroid(x_music_array, sr = sr_sample_rate))

    row.append(spectral_centroids)
    row.append(spectral_centroids_var)


    # 4/ spectral_rolloff (par_3)
    spectral_rolloff = np.mean(librosa.feature.spectral_rolloff(x_music_array, sr = sr_sample_rate))
    spectral_rolloff_var = np.var(librosa.feature.spectral_rolloff(x_music_array, sr = sr_sample_rate))

    row.append(spectral_rolloff)
    row.append(spectral_rolloff_var)

    # 5/ mfccs (par_4)
    mfccs = np.mean(librosa.feature.mfcc(x_music_array, sr = sr_sample_rate))
    mfccs_var = np.var(librosa.feature.mfcc(x_music_array, sr = sr_sample_rate))
    row.append(mfccs)
    row.append(mfccs_var)


    # 6/ chroma_frequencies (par_5)
    chromogram = np.mean(librosa.feature.chroma_stft(x_music_array, sr = sr_sample_rate))
    chromogram_var = np.var(librosa.feature.chroma_stft(x_music_array, sr = sr_sample_rate))
    row.append(chromogram)
    row.append(chromogram_var)
    

    arr = numpy.array(row)
    #print(arr)
    #print(arr.reshape(1, -1))
    
    arr_X = scaler.transform(arr.reshape(1, -1)) # X values of a given tune
    
    predict_new_Y = model.predict(arr_X)
    round_off_values = np.round_(predict_new_Y)
        
    
    return di[np.argmax(round_off_values)]
    
print("Here is the genre of the song")
print(showing_genre_of_a_tune("blues-Copy1.00009.wav")) 
# Enter the audio file name above
```
